```python
import os
import pandas as pd

data_dir = ".\data"

data_files = [f for f in os.listdir(data_dir) if f.endswith(".sas7bdat")]

dataframes = {file: pd.read_sas(os.path.join(data_dir, file)) for file in data_files}

for file, df in dataframes.items():
    print(f"\n----- {file} -----")
    print(df.info())
    print("\nSample Data:")
    display(df.head())
```
